```python
import csv
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.ticker as ticker
import numpy as np
```

```python
# data = []
file1 = 'C:/Users/s174370/magfield-prediction/wandb_plots/wandb_scalar-pot_and_main2D.csv'
# with open(file1, 'r') as file:
#     csv_reader = csv.reader(file)
#     for row in csv_reader:
#         data.append(row)


file2 = 'C:/Users/s174370/magfield-prediction/wandb_plots\wandb_SP_gauge.csv'
df = pd.read_csv(file1)
df2 = pd.read_csv(file2)

print(df.columns)
df2.columns



```

```python

x = df['Step']
y1 = df['main_2D - ae']
y2 = df['scalar_potential - ae']
y3 = df2['vague-yogurt-24 - ae']
```

```python
# fig = plt.figure(figsize=(7,4))

# plt.rcParams['font.family'] = 'Calibri'

fig,ax = plt.subplots(figsize=(7,4))
plt.yscale('log')
ax.plot(x, y1, label='Main 2D')
ax.plot(x, y2, label='SP')
# ax.plot(x, y3, label='SP with gauge')



# Define the formatter function
def mini_tesla_formatter(x, pos):
    return f"{x*1000:.2f}"  # Convert to mini Tesla and format the value

# Apply the formatter to the Y-axis
ax.yaxis.set_major_formatter(ticker.FuncFormatter(mini_tesla_formatter))
plt.rcParams['font.size'] = 10
plt.xlabel('Steps')
plt.ylabel('Absolute Error [mT], log scale')
plt.title(r'Absolute error, $L_{match}$')
plt.legend()
# plt.grid(True)
plt.savefig('main2D_vs_SP_log.pdf')
plt.show()
```
